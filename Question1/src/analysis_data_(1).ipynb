```python
import pandas as pd
df = pd.read_csv('/content/data_clean/clean_frailty.csv')
df=df.drop('Unnamed: 0',axis=1) #Dropping the column named 'Unnamed: 0' from the DataFrame.
df.head()
```

```python
import matplotlib as plt
import seaborn as sns
#importing the required libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1: Bar plot of Height vs. Weight
plt.figure(figsize=(10, 6))
sns.barplot(x='height', y='weight', data=df, palette='coolwarm')
plt.xlabel('Height (in)', fontsize=12)
plt.ylabel('Weight (lbs)', fontsize=12)
plt.title('Bar plot of Height vs. Weight', fontsize=14)
plt.show()

# 2: KDE plot of Age (Density plot)
plt.figure(figsize=(10, 6))
sns.kdeplot(df['age'], shade=True, color='purple')
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('KDE plot of Age', fontsize=14)
plt.show()

# 3: Swarm plot of Grip Strength grouped by Frailty
plt.figure(figsize=(10, 6))
sns.swarmplot(x='frailty', y='grip_strength', data=df, palette='Set2', size=7)
plt.xlabel('Frailty', fontsize=12)
plt.ylabel('Grip Strength (kg)', fontsize=12)
plt.title('Swarm plot of Grip Strength (Grouped by Frailty)', fontsize=14)
plt.show()

# 4: Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['height'], df['weight'])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Scatter plot of height vs. weight')
plt.grid(True)
plt.show()

# 5: Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='frailty', y='grip_strength')
plt.xlabel('frailty')
plt.ylabel('grip_strength')
plt.title('Box plot of Grip strength (Grouped by Frailty)')
plt.grid(True)
plt.show()
```
