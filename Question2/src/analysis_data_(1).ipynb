**Data Visualization & Analysis Stage: Created five different visualizations to explore the dataset and provided insights.**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data_clean/StudentsPerformance (2).csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# 1. Joint Plot
sns.jointplot(data=df, x='reading score', y='writing score', kind='scatter')
plt.xlabel('reading score')
plt.ylabel('writing Score')
plt.title('Joint Plot of Reading vs. Writing Scores')
plt.show()

# 2. Heatmap
corr = df[['math score', 'reading score', 'writing score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='white', fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()

# 3. Box Plot with Swarm Overlay
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='parental level of education', y='math score')
sns.swarmplot(data=df, x='parental level of education', y='math score', color='blue', alpha=0.7)
plt.xlabel('parental level of education')
plt.ylabel('math score')
plt.title('Box Plot with Swarm Overlay of Math Scores by Parental Level of Education')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 4. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='gender', y='math score')
plt.xlabel('gender')
plt.ylabel('math score')
plt.title('Violin Plot of Math Scores by Gender')
plt.grid(True)
plt.show()
```

```python
# 5. PairGrid
g = sns.PairGrid(df[['math score', 'reading score', 'writing score']])
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.suptitle('PairGrid of Math, Reading, and Writing Scores')
plt.show()
```

```python
# 6. Swarm Plot
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df, x='parental level of education', y='reading score')
plt.xlabel('parental level of education')
plt.ylabel('reading score')
plt.title('Swarm Plot of Reading Scores by Parental Level of Education')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
